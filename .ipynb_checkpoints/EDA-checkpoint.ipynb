## Quick Viz for Categorical

```python
for i in df_cat.columns:
    plt.figure(figsize=(10,5))
    cat_num = df_cat[i].value_counts()
    print("graph for %s: total = %d" %(i,len(cat_num)))
    sns.barplot(x=cat_num.index,y=cat_num)
    plt.xticks(rotation=90)
    plt.show()
```

## Pivot Table

```python
#by state
pd.set_option('display.max_rows',None)
pd.pivot_table(df, index=['job_state','seniority'],values='avg_salary')
```

```python
#number of job opening by state
pd.set_option('display.max_rows',None)
pd.pivot_table(df, index=['job_state','seniority'],values='avg_salary',aggfunc='count').sort_values('job_state',ascending=False)
```

```python
#salary by rating
cat_salary = ['Rating','Sector','Size','Type of ownership']
for i in cat_salary:
    pd.pivot_table(df, index=[i],values='avg_salary').sort_values(by='avg_salary',ascending=False).plot(kind='bar')
```

# Word Cloud of Job Desc

```python
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
words = " ".join(df['Job Description'])

def punctuation_stop(text):
    """remove punctuation and stop words"""
    filtered = []
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    for w in word_tokens:
        if w not in stop_words and w.isalpha():
            filtered.append(w.lower())
    return filtered


words_filtered = punctuation_stop(words)

text = " ".join([ele for ele in words_filtered])

wc= WordCloud(background_color="white", random_state=1,stopwords=STOPWORDS, max_words = 2000, width =800, height = 1500)
wc.generate(text)

plt.figure(figsize=[10,10])
plt.imshow(wc,interpolation="bilinear")
plt.axis('off')
plt.show()
```
